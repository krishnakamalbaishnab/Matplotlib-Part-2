# the long version of plt.plot()

```python
# Creating a plot with the OO verison, confusing way first
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# Confusing #2
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# Easier and more robust going forward (Most Usedf and recomended)
fig, ax = plt.subplots()
ax.plot(x, y);
```
