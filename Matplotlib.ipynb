# the long version of plt.plot()

```python
# Creating a plot with the OO verison, confusing way first
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# Confusing #2
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# Easier and more robust going forward (Most Usedf and recomended)
fig, ax = plt.subplots()
ax.plot(x, y);
```

## Matplotlib example workflow

```python
 # 1. Import the matplotlib libray 
    
%matplotlib inline
import matplotlib.pyplot as plt

#1.Prepare data
 
x = [1,2,3,4,5]
y = [10,20,30,40,50]

#2. Setup plot 

fig, ax = plt.subplots(figsize = (10,10))


#3. Plot Data

ax.plot(x,y)

# Customise plot 

ax.set(title  = "Sample Plot",
       xlabel= "X -Axis",
       ylabel="Y - Axis")

#5 Save  and show the plot

#fig.savefig("../images/sample-plot.png")
```

# Line

Line is the default type of visualization in Matplotlib. Usually, unless specified otherwise, your plots will start out as lines.

```python
# Create an array
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# The default plot is line
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Need to recreate our figure and axis instances when we want a new figure
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```
